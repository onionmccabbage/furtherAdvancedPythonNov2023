#### Using Basic Numpy


```python
# numpy offers additional data structures beyond those available in Python
# array range (arange)
# start, stop-before, step
n = np.arange(3.0,12.5,0.5) # all the members must be of the same data type
n
n.dtype
```

```python
m = np.arange('a', 'b', 'c') # fix this <------------
m # throws an exception
```

```python
# we can use many-dimensions in our arrays
# we have a mutable collection of one data type
d = np.random.randn(9,3) # we have an array of 9 by 3 random values
# d = np.random.randn(900,300) # we have an array of 900 by 300 random values
d.dtype
d.shape # (9, 3)
d
```

```python
# multi dimensional arrays
e =np.random.randn(2,3,2,5,2,3)
e # e.shape, e.dtype
```

```python
# Creating arrays
f = np.zeros( (3,6) ) # careful: n.empty() will NOT zero-out the memory
f.dtype # we have floats
```
